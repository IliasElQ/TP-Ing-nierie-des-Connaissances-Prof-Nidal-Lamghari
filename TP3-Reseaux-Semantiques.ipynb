## Exercice 1 : HÃ©ritage Multiple - Le cas de la Chauve-souris

### Ã‰noncÃ©

**Faits :**
- Oiseau âŸ¶ *isa* âŸ¶ Animal ; Oiseau âŸ¶ *possÃ¨de* âŸ¶ peut_voler
- MammifÃ¨re âŸ¶ *isa* âŸ¶ Animal ; MammifÃ¨re âŸ¶ *possÃ¨de* âŸ¶ allaite
- Chauve-souris âŸ¶ *isa* âŸ¶ MammifÃ¨re **et** âŸ¶ *isa* âŸ¶ Oiseau
- Chauve-souris âŸ¶ *possÃ¨de* âŸ¶ Â¬pond_oeufs

### Questions :
1. ReprÃ©senter le rÃ©seau sÃ©mantique
2. Lister les propriÃ©tÃ©s hÃ©ritÃ©es par "Chauve-souris"
3. Identifier les conflits potentiels

```python
# Exercice 1 : CrÃ©ation du rÃ©seau sÃ©mantique pour Chauve-souris

def create_bat_network():
    """
    CrÃ©e un rÃ©seau sÃ©mantique illustrant l'hÃ©ritage multiple de la Chauve-souris
    """
    G = nx.DiGraph()
    
    # Ajout des nÅ“uds (concepts)
    nodes = {
        'Animal': {'type': 'classe', 'level': 0},
        'Oiseau': {'type': 'classe', 'level': 1},
        'MammifÃ¨re': {'type': 'classe', 'level': 1},
        'Chauve-souris': {'type': 'instance', 'level': 2},
        'peut_voler': {'type': 'propriÃ©tÃ©', 'level': 1.5},
        'allaite': {'type': 'propriÃ©tÃ©', 'level': 1.5},
        'Â¬pond_oeufs': {'type': 'propriÃ©tÃ©', 'level': 2.5}
    }
    
    for node, attrs in nodes.items():
        G.add_node(node, **attrs)
    
    # Ajout des relations (arÃªtes)
    edges = [
        ('Oiseau', 'Animal', 'isa'),
        ('MammifÃ¨re', 'Animal', 'isa'),
        ('Chauve-souris', 'Oiseau', 'isa'),
        ('Chauve-souris', 'MammifÃ¨re', 'isa'),
        ('Oiseau', 'peut_voler', 'possÃ¨de'),
        ('MammifÃ¨re', 'allaite', 'possÃ¨de'),
        ('Chauve-souris', 'Â¬pond_oeufs', 'possÃ¨de')
    ]
    
    for src, dst, rel_type in edges:
        G.add_edge(src, dst, relation=rel_type)
    
    return G, nodes

# CrÃ©ation du graphe
G_bat, nodes_bat = create_bat_network()

# Visualisation
plt.figure(figsize=(14, 10))
pos = {}

# Positionnement manuel pour une meilleure lisibilitÃ©
pos['Animal'] = (0, 3)
pos['Oiseau'] = (-2, 2)
pos['MammifÃ¨re'] = (2, 2)
pos['peut_voler'] = (-3.5, 1.5)
pos['allaite'] = (3.5, 1.5)
pos['Chauve-souris'] = (0, 0.5)
pos['Â¬pond_oeufs'] = (0, -0.5)

# Couleurs selon le type
colors = []
for node in G_bat.nodes():
    if nodes_bat[node]['type'] == 'classe':
        colors.append('#87CEEB')  # Bleu clair pour classes
    elif nodes_bat[node]['type'] == 'propriÃ©tÃ©':
        colors.append('#98FB98')  # Vert clair pour propriÃ©tÃ©s
    else:
        colors.append('#FFB6C1')  # Rose pour instances

# Dessin des nÅ“uds
nx.draw_networkx_nodes(G_bat, pos, node_color=colors, node_size=3000, 
                       node_shape='o', alpha=0.9, linewidths=2, edgecolors='black')

# Dessin des labels
nx.draw_networkx_labels(G_bat, pos, font_size=10, font_weight='bold')

# Dessin des arÃªtes avec diffÃ©rents styles selon le type
for (u, v, d) in G_bat.edges(data=True):
    rel_type = d['relation']
    if rel_type == 'isa':
        nx.draw_networkx_edges(G_bat, pos, [(u, v)], edge_color='blue', 
                              width=2, arrows=True, arrowsize=20, 
                              arrowstyle='->', connectionstyle='arc3,rad=0.1')
    else:  # possÃ¨de
        nx.draw_networkx_edges(G_bat, pos, [(u, v)], edge_color='green', 
                              width=2, arrows=True, arrowsize=20, 
                              arrowstyle='->', style='dashed', connectionstyle='arc3,rad=0.1')

# LÃ©gende
from matplotlib.lines import Line2D
legend_elements = [
    Line2D([0], [0], color='blue', lw=2, label='isa (hÃ©ritage)'),
    Line2D([0], [0], color='green', lw=2, linestyle='--', label='possÃ¨de (propriÃ©tÃ©)'),
    Line2D([0], [0], marker='o', color='w', markerfacecolor='#87CEEB', 
           markersize=10, label='Classe', markeredgecolor='black'),
    Line2D([0], [0], marker='o', color='w', markerfacecolor='#98FB98', 
           markersize=10, label='PropriÃ©tÃ©', markeredgecolor='black'),
    Line2D([0], [0], marker='o', color='w', markerfacecolor='#FFB6C1', 
           markersize=10, label='Instance', markeredgecolor='black')
]
plt.legend(handles=legend_elements, loc='upper left', fontsize=10)

plt.title("Exercice 1 : RÃ©seau SÃ©mantique - HÃ©ritage Multiple (Chauve-souris)", 
          fontsize=14, fontweight='bold')
plt.axis('off')
plt.tight_layout()
plt.show()

print("âœ“ RÃ©seau sÃ©mantique gÃ©nÃ©rÃ© avec succÃ¨s!")
```

### RÃ©ponse 1.3 : Analyse des Conflits

Deux types de conflits apparaissent dans ce rÃ©seau :

```python
# Analyse des conflits d'hÃ©ritage

def analyze_conflicts():
    """
    Identifie et analyse les conflits dans l'hÃ©ritage de la Chauve-souris
    """
    print("=" * 70)
    print("ANALYSE DES CONFLITS D'HÃ‰RITAGE")
    print("=" * 70)
    
    conflicts = [
        {
            'type': 'Conflit Taxonomique',
            'description': 'Double hÃ©ritage : Oiseau âˆ§ MammifÃ¨re',
            'explication': """
            Dans une ontologie rÃ©aliste, les classes Oiseau et MammifÃ¨re 
            sont normalement disjointes (un animal ne peut pas Ãªtre les deux).
            La Chauve-souris crÃ©e une exception Ã  cette rÃ¨gle.""",
            'resolution': 'Accepter l\'hÃ©ritage multiple exceptionnel',
            'severity': 'âš ï¸ MODÃ‰RÃ‰'
        },
        {
            'type': 'Conflit de PropriÃ©tÃ©s',
            'description': 'pond_oeufs vs Â¬pond_oeufs',
            'explication': """
            Si on ajoute la rÃ¨gle par dÃ©faut : Oiseau âŸ¶ pond_oeufs
            Conflit avec : Chauve-souris âŸ¶ Â¬pond_oeufs
            
            RÃ¨gle hÃ©ritÃ©e (gÃ©nÃ©rale) : pond_oeufs (de Oiseau)
            RÃ¨gle spÃ©cifique (locale) : Â¬pond_oeufs (de Chauve-souris)""",
            'resolution': 'Principe de SPÃ‰CIFICITÃ‰ : la rÃ¨gle la plus spÃ©cifique prime',
            'severity': 'âœ“ RÃ‰SOLU'
        }
    ]
    
    for i, conflict in enumerate(conflicts, 1):
        print(f"\nðŸ”´ CONFLIT {i}: {conflict['type']}")
        print(f"   Description: {conflict['description']}")
        print(f"   Explication: {conflict['explication']}")
        print(f"   RÃ©solution: {conflict['resolution']}")
        print(f"   Statut: {conflict['severity']}")
        print("-" * 70)
    
    # Visualisation de la rÃ©solution du conflit
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 5))
    
    # Graphique 1: Conflit taxonomique
    ax1.text(0.5, 0.8, 'CONFLIT TAXONOMIQUE', ha='center', fontsize=14, 
             fontweight='bold', transform=ax1.transAxes)
    ax1.text(0.5, 0.6, 'Oiseau âˆ§ MammifÃ¨re = ?', ha='center', fontsize=12, 
             transform=ax1.transAxes, bbox=dict(boxstyle='round', facecolor='yellow', alpha=0.5))
    ax1.text(0.5, 0.4, 'â†“', ha='center', fontsize=20, transform=ax1.transAxes)
    ax1.text(0.5, 0.25, 'Chauve-souris\n(Exception)', ha='center', fontsize=11, 
             transform=ax1.transAxes, bbox=dict(boxstyle='round', facecolor='lightcoral', alpha=0.7))
    ax1.axis('off')
    ax1.set_title('Classes Disjointes', fontsize=12, fontweight='bold')
    
    # Graphique 2: RÃ©solution par spÃ©cificitÃ©
    ax2.text(0.5, 0.85, 'RÃ‰SOLUTION PAR SPÃ‰CIFICITÃ‰', ha='center', fontsize=14, 
             fontweight='bold', transform=ax2.transAxes)
    ax2.text(0.5, 0.65, 'Oiseau âŸ¶ pond_oeufs', ha='center', fontsize=10, 
             transform=ax2.transAxes, bbox=dict(boxstyle='round', facecolor='lightblue', alpha=0.5))
    ax2.text(0.5, 0.5, 'VS', ha='center', fontsize=12, fontweight='bold', 
             transform=ax2.transAxes, color='red')
    ax2.text(0.5, 0.35, 'Chauve-souris âŸ¶ Â¬pond_oeufs', ha='center', fontsize=10, 
             transform=ax2.transAxes, bbox=dict(boxstyle='round', facecolor='lightgreen', alpha=0.5))
    ax2.text(0.5, 0.15, 'âœ“ Plus spÃ©cifique GAGNE!', ha='center', fontsize=11, 
             fontweight='bold', color='green', transform=ax2.transAxes)
    ax2.axis('off')
    ax2.set_title('Principe de Non-MonotonicitÃ©', fontsize=12, fontweight='bold')
    
    plt.tight_layout()
    plt.show()
    
    print("\n" + "=" * 70)
    print("CONCLUSION:")
    print("La Chauve-souris hÃ©rite de 'peut_voler' et 'allaite'")
    print("La propriÃ©tÃ© 'Â¬pond_oeufs' est une exception spÃ©cifique qui prime")
    print("=" * 70)

# ExÃ©cution de l'analyse
analyze_conflicts()
```

## Exercice 2 : Raisonnement par HÃ©ritage - Le Dauphin nage

### Ã‰noncÃ©

**RÃ¨gles et Faits :**
- Poisson âŸ¶ vit_dans âŸ¶ Eau
- Animal âˆ§ vit_dans(Eau) âŸ¶ peut_nager
- Dauphin âŸ¶ *isa* âŸ¶ MammifÃ¨re ; MammifÃ¨re âŸ¶ *isa* âŸ¶ Animal
- Dauphin âŸ¶ vit_dans âŸ¶ Eau

### Objectif :
DÃ©montrer par le rÃ©seau sÃ©mantique et l'hÃ©ritage que : **Dauphin peut_nager**

```python
# Exercice 2 : DÃ©monstration - Dauphin peut nager

def create_dolphin_network():
    """
    CrÃ©e un rÃ©seau sÃ©mantique pour dÃ©montrer que le Dauphin peut nager
    """
    G = nx.DiGraph()
    
    # NÅ“uds
    nodes = {
        'Animal': {'type': 'classe', 'pos': (0, 3)},
        'MammifÃ¨re': {'type': 'classe', 'pos': (0, 2)},
        'Dauphin': {'type': 'instance', 'pos': (0, 1)},
        'Eau': {'type': 'propriÃ©tÃ©', 'pos': (2, 1)},
        'peut_nager': {'type': 'propriÃ©tÃ© dÃ©rivÃ©e', 'pos': (2, 0)},
        'Poisson': {'type': 'classe', 'pos': (-2, 2)}
    }
    
    for node, attrs in nodes.items():
        G.add_node(node, **attrs)
    
    # ArÃªtes avec types de relations
    edges = [
        ('MammifÃ¨re', 'Animal', 'isa', 'blue'),
        ('Dauphin', 'MammifÃ¨re', 'isa', 'blue'),
        ('Dauphin', 'Eau', 'vit_dans', 'green'),
        ('Poisson', 'Eau', 'vit_dans', 'orange')
    ]
    
    for src, dst, rel, color in edges:
        G.add_edge(src, dst, relation=rel, color=color)
    
    return G, nodes

# CrÃ©ation du rÃ©seau
G_dolphin, nodes_dolphin = create_dolphin_network()

# Visualisation
fig, ax = plt.subplots(figsize=(14, 10))

pos = {node: attrs['pos'] for node, attrs in nodes_dolphin.items()}

# Couleurs des nÅ“uds
node_colors = {
    'classe': '#87CEEB',
    'instance': '#FFB6C1',
    'propriÃ©tÃ©': '#98FB98',
    'propriÃ©tÃ© dÃ©rivÃ©e': '#FFD700'
}

colors = [node_colors[nodes_dolphin[node]['type']] for node in G_dolphin.nodes()]

# Dessin des nÅ“uds
nx.draw_networkx_nodes(G_dolphin, pos, node_color=colors, node_size=3500,
                       node_shape='o', alpha=0.9, linewidths=2.5, edgecolors='black')

# Labels
nx.draw_networkx_labels(G_dolphin, pos, font_size=11, font_weight='bold')

# Dessin des arÃªtes avec couleurs
for (u, v, d) in G_dolphin.edges(data=True):
    color = d['color']
    nx.draw_networkx_edges(G_dolphin, pos, [(u, v)], edge_color=color,
                          width=2.5, arrows=True, arrowsize=25,
                          arrowstyle='->', connectionstyle='arc3,rad=0.1')

# Ajout de la rÃ¨gle d'infÃ©rence
rule_box = FancyBboxPatch((1.2, -0.8), 2.8, 1.2, boxstyle="round,pad=0.1",
                          edgecolor='red', facecolor='lightyellow', linewidth=2)
ax.add_patch(rule_box)
ax.text(2.6, 0.2, 'RÃˆGLE D\'INFÃ‰RENCE:', fontsize=10, fontweight='bold',
        ha='center', color='red')
ax.text(2.6, -0.1, 'Animal âˆ§ vit_dans(Eau)', fontsize=9, ha='center')
ax.text(2.6, -0.35, 'âŸ¹', fontsize=14, ha='center', fontweight='bold')
ax.text(2.6, -0.6, 'peut_nager', fontsize=9, ha='center',
        bbox=dict(boxstyle='round', facecolor='#FFD700', alpha=0.7))

# FlÃ¨ches de dÃ©duction
ax.annotate('', xy=(2, 0), xytext=(0.2, 1),
            arrowprops=dict(arrowstyle='->', lw=2, color='red', linestyle='dashed'))
ax.annotate('', xy=(2, 0), xytext=(1.8, 1),
            arrowprops=dict(arrowstyle='->', lw=2, color='red', linestyle='dashed'))

# LÃ©gende
from matplotlib.lines import Line2D
legend_elements = [
    Line2D([0], [0], color='blue', lw=2, label='isa (hÃ©ritage)'),
    Line2D([0], [0], color='green', lw=2, label='vit_dans'),
    Line2D([0], [0], color='red', lw=2, linestyle='--', label='infÃ©rence'),
    Line2D([0], [0], marker='o', color='w', markerfacecolor='#87CEEB',
           markersize=10, label='Classe', markeredgecolor='black'),
    Line2D([0], [0], marker='o', color='w', markerfacecolor='#FFB6C1',
           markersize=10, label='Instance', markeredgecolor='black'),
    Line2D([0], [0], marker='o', color='w', markerfacecolor='#FFD700',
           markersize=10, label='PropriÃ©tÃ© dÃ©rivÃ©e', markeredgecolor='black')
]
ax.legend(handles=legend_elements, loc='upper left', fontsize=10)

plt.title("Exercice 2 : DÃ©monstration par RÃ©seau SÃ©mantique\nDauphin peut_nager",
          fontsize=14, fontweight='bold')
plt.axis('off')
plt.xlim(-3.5, 4.5)
plt.ylim(-1.5, 3.5)
plt.tight_layout()
plt.show()

print("âœ“ RÃ©seau sÃ©mantique gÃ©nÃ©rÃ©!")
```

## Exercice 3 : RÃ©seau AcadÃ©mique avec Contraintes

### Ã‰noncÃ©

**Structure Ã  reprÃ©senter :**
- Personne âŸ¶ aAge: â„• (attribut numÃ©rique)
- Ã‰tudiant âŸ¶ *isa* âŸ¶ Personne ; Ã‰tudiant âŸ¶ Ã©tudie âŸ¶ Domaine
- Enseignant âŸ¶ *isa* âŸ¶ Personne ; Enseignant âŸ¶ enseigne âŸ¶ Domaine
- Professeur âŸ¶ *isa* âŸ¶ Enseignant âˆ§ Chercheur
- **Contrainte** : pour un mÃªme cours, domaine_enseignÃ© = domaine_Ã©tudiÃ©

### Questions :
1. ReprÃ©senter le rÃ©seau complet

## Exercice 4 : Propagation d'Activation

### Ã‰noncÃ©

**RÃ©seau donnÃ© :**
- Chien â”€isaâ†’ Animal
- Chat â”€isaâ†’ Animal
- Animal â”€isaâ†’ ÃŠtreVivant
- Chien â”€liÃ©Ã€â†’ Os
- Chat â”€liÃ©Ã€â†’ Lait
- Os â”€typeâ†’ Nourriture

**ParamÃ¨tres de propagation :**
- Niveau maximum : 3
- Facteur de propagation : 0.5
- Activation initiale : Aâ‚€(Chien) = 1.00

### Objectif :
Calculer les niveaux d'activation de tous les nÅ“uds

```python
# Exercice 4 : Propagation d'activation

def create_activation_network():
    """
    CrÃ©e le rÃ©seau pour la propagation d'activation
    """
    G = nx.DiGraph()
    
    # NÅ“uds
    nodes = ['Chien', 'Chat', 'Animal', 'ÃŠtreVivant', 'Os', 'Lait', 'Nourriture']
    G.add_nodes_from(nodes)
    
    # ArÃªtes
    edges = [
        ('Chien', 'Animal'),
        ('Chat', 'Animal'),
        ('Animal', 'ÃŠtreVivant'),
        ('Chien', 'Os'),
        ('Chat', 'Lait'),
        ('Os', 'Nourriture')
    ]
    G.add_edges_from(edges)
    
    return G

def propagate_activation(G, initial_node, max_level=3, factor=0.5):
    """
    Simule la propagation d'activation dans le rÃ©seau
    
    Args:
        G: graphe NetworkX
        initial_node: nÅ“ud de dÃ©part
        max_level: nombre maximum de niveaux de propagation
        factor: facteur de dÃ©croissance Ã  chaque niveau
    
    Returns:
        dict: activations par nÅ“ud et par niveau
    """
    # Initialisation
    activations = {node: 0.0 for node in G.nodes()}
    activations[initial_node] = 1.0
    
    # Historique par niveau
    level_history = {0: {initial_node: 1.0}}
    
    # Propagation niveau par niveau
    current_active = {initial_node: 1.0}
    
    for level in range(1, max_level + 1):
        next_active = {}
        
        for node, activation in current_active.items():
            # Propage vers tous les successeurs
            for successor in G.successors(node):
                new_activation = activation * factor
                # Prend le maximum si le nÅ“ud est dÃ©jÃ  activÃ©
                if successor in next_active:
                    next_active[successor] = max(next_active[successor], new_activation)
                else:
                    next_active[successor] = new_activation
                
                # Met Ã  jour l'activation globale
                activations[successor] = max(activations[successor], new_activation)
        
        level_history[level] = next_active.copy()
        current_active = next_active
        
        if not current_active:
            break
    
    return activations, level_history

# CrÃ©ation du rÃ©seau
G_activation = create_activation_network()

# Propagation depuis "Chien"
activations, level_history = propagate_activation(G_activation, 'Chien', max_level=3, factor=0.5)

# Affichage des rÃ©sultats
print("=" * 70)
print("PROPAGATION D'ACTIVATION")
print("=" * 70)
print(f"\nNÅ“ud initial: Chien (activation = 1.00)")
print(f"Facteur de propagation: 0.5")
print(f"Niveaux maximum: 3")
print("\n" + "=" * 70)

for level in sorted(level_history.keys()):
    print(f"\nðŸ“ NIVEAU {level}:")
    if level_history[level]:
        for node, activation in sorted(level_history[level].items(), 
                                       key=lambda x: x[1], reverse=True):
            print(f"   {node:15s} = {activation:.3f}")
    else:
        print("   (aucune nouvelle activation)")

print("\n" + "=" * 70)
print("BILAN FINAL DES ACTIVATIONS:")
print("=" * 70)
for node, activation in sorted(activations.items(), key=lambda x: x[1], reverse=True):
    if activation > 0:
        print(f"   {node:15s} = {activation:.3f}")

print("\nâœ“ Propagation terminÃ©e!")
```

```python
# Analyse du raisonnement diagnostic

def diagnostic_reasoning(symptoms):
    """
    Raisonnement diagnostic basÃ© sur les symptÃ´mes
    """
    print(f"{'='*80}")
    print(f"ðŸ¥ ANALYSE DIAGNOSTIQUE - SymptÃ´mes du patient : {', '.join(symptoms)}")
    print(f"{'='*80}\n")
    
    # Base de connaissance
    maladies = {
        'Grippe': {
            'symptomes_directs': ['FiÃ¨vre', 'Toux'],
            'classe': 'MaladieRespiratoire'
        },
        'Pneumonie': {
            'symptomes_directs': ['FiÃ¨vre', 'DouleurPoitrine'],
            'classe': 'MaladieRespiratoire'
        }
    }
    
    symptomes_classe = {
        'MaladieRespiratoire': ['Toux']
    }
    
    diagnostics = []
    
    for maladie, info in maladies.items():
        print(f"\nðŸ“‹ Test : {maladie}")
        print(f"   Classe : {info['classe']}")
        print(f"   SymptÃ´mes directs associÃ©s : {info['symptomes_directs']}")
        
        # SymptÃ´mes hÃ©ritÃ©s de la classe
        symptomes_herites = symptomes_classe.get(info['classe'], [])
        if symptomes_herites:
            print(f"   SymptÃ´mes hÃ©ritÃ©s (classe) : {symptomes_herites}")
        
        # Tous les symptÃ´mes possibles
        all_symptoms = set(info['symptomes_directs'] + symptomes_herites)
        
        # VÃ©rification
        matched = [s for s in symptoms if s in all_symptoms]
        
        if matched:
            score = len(matched) / len(symptoms)
            diagnostics.append({
                'maladie': maladie,
                'symptomes_matches': matched,
                'score': score,
                'type': 'exact' if score == 1.0 else 'partiel'
            })
            
            print(f"   âœ“ Correspondance : {matched} ({len(matched)}/{len(symptoms)})")
            print(f"   Score : {score*100:.0f}%")
            
            # Explication du raisonnement
            print(f"\n   ðŸ’¡ Raisonnement :")
            for symptom in matched:
                if symptom in info['symptomes_directs']:
                    print(f"      â€¢ {symptom} â†’ hasSymptom({maladie}, {symptom})")
                elif symptom in symptomes_herites:
                    print(f"      â€¢ {symptom} â†’ hasSymptom(MaladieRespiratoire, {symptom})")
                    print(f"                 + isa({maladie}, MaladieRespiratoire)")
                    print(f"                 â†’ {maladie} HÃ‰RITE de ce symptÃ´me")
        else:
            print(f"   âœ— Aucune correspondance")
    
    # RÃ©sultat final
    print(f"\n{'='*80}")
    print(f"ðŸŽ¯ CONCLUSION DIAGNOSTIQUE :")
    print(f"{'='*80}\n")
    
    if diagnostics:
        # Tri par score dÃ©croissant
        diagnostics.sort(key=lambda x: x['score'], reverse=True)
        
        for i, diag in enumerate(diagnostics, 1):
            emoji = 'ðŸ¥‡' if i == 1 else 'ðŸ¥ˆ' if i == 2 else 'ðŸ¥‰'
            print(f"{emoji} {diag['maladie']} - Score: {diag['score']*100:.0f}%")
            print(f"   Type: {diag['type'].upper()}")
            print(f"   SymptÃ´mes identifiÃ©s: {', '.join(diag['symptomes_matches'])}")
            print()
        
        print(f"ðŸ”¬ Recommandation :")
        best = diagnostics[0]
        if best['score'] == 1.0:
            print(f"   Diagnostic probable : {best['maladie']} (correspondance totale)")
        else:
            print(f"   Examens complÃ©mentaires recommandÃ©s pour diffÃ©rencier :")
            for diag in diagnostics:
                print(f"   - {diag['maladie']}")
    else:
        print("âŒ Aucune maladie identifiÃ©e dans la base de connaissances")
    
    print(f"\n{'='*80}")

# Test avec les symptÃ´mes du patient
symptomes_patient = ['FiÃ¨vre', 'Toux']
diagnostic_reasoning(symptomes_patient)
```
